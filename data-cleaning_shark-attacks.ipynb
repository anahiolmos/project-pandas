```python
def clean_columns(df):
    
    '''
    Clean columns names:
        Input: columns names.
        Return: modified column names in dataframe (df).
    '''
    
    column_clean = []
    for column in df.columns:
        column = column.strip().lower()
        column = column.replace(' ','_')
        column = column.replace('.','_')
        column = column.replace(':','_')
        column_clean.append(column)
        
    df.columns = column_clean
    
    return df.columns
```

### agrupando actividades únicas (value_count = 1) como 'others':

```python
df['activity'].value_counts()
```

```python
# aplicando groupby a las actividades individuales
group_singles_act = list(df.groupby('activity').filter(lambda x: len(x) == 1).activity)

# groupby en un diccionario que llame 'others' a estas actividades individuales
dic_singles_act = {'others': group_singles_act}

# función que agrupa las actividades individuales
def single_activity(activity):
    
    '''
    Iterates each single activity and groups it into 'others':
        Input: dictionary with activities grouped in 'others'.
        Return: activity.
    '''
    
    for act, count in dic_singles_act.items():
        if activity in count:
            return act
        else:
            return activity

df['activity'] = df['activity'].apply(single_activity)
```

```python
# Función para codificar el 'tipo de ataque'

def encode_type_attack(attack: str) -> int:
    
    """
  Recieves a value (type attack by shark) and evaluates it to a number.
      Return: int
    """
    
    if attack == 'Unprovoked':
        return 1
    elif attack == 'Provoked':
        return 2
    elif attack == 'Boat':
        return 3
    elif attack == 'Boating':
        return 4
    elif attack == 'Sea Disaster':
        return 5
    elif attack == 'unknown':
        return 0
    elif attack == 'Invalid':
        return 0
```

```python
df['type_encoded'] = df['type'].apply(encode_type_attack)
```

```python
df['age'].fillna('unkown', inplace=True)

def clean_age(df):
    
    '''
    Clean age names:
        Input: 'age' names.
        Return: modified 'age' names in the whole dataframe.
    '''
    
    df['age'] = df['age'].astype(str)

    ages = []
    for age in df['age']:
        age = age.strip().lower()
        age = age.replace('?','')
        age = age.replace(' &',',')
        age = age.replace('s','')
        age = age.replace("'s","")
        age = age.replace('>','')
        age = age.replace('[','')
        age = age.replace(']','')
        age = age.replace('(','')
        age = age.replace(')','')
        age = age.replace('"','')
        age = age.replace("''",'')
        ages.append(age)
                             
    df['age'] = ages
    
    return df['age']

df['age'].replace(' or.*,*', '', regex=True, inplace=True)
df['age'].replace(' to.*,*', '', regex=True, inplace=True)
```
